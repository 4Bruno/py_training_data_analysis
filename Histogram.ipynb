```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.array(np.random.random(size=(4,4)) * 10, dtype='uint32')
```

```python
a
```

```python
def get_hist_bins_from_matrix(m):
    return np.histogram(m)[1]
```

```python
plt.hist(a.flat,bins = get_hist_bins_from_matrix(a))
plt.show()
```

```python
# histogram 2d
# take 2 arrays to represente x/y axis
h2d = np.histogram2d(a[0],a[1], bins = 4)
print("np.histogram2d returns a list of 3 arrays with information about X/Y axis bins and density array")
print("This is the list of bins for X-axis")
print(h2d[1])
print("This is the list of bins for Y-axis")
print(h2d[2])
print("Density matrix defines density for x-axis bins as rows")
print(h2d[0])
print("Let's take a look at first row, the values indicate which bins in the Y-axis have any density")
print(h2d[0][0])
print("This means that within bin[0] of X-axis, there is density in bin[2] of Y-axis")
print("------------------------------------------------------------------------------------------------")
print("(If you rotate the density matrix counter clock-wise 45d you get the traditional plotted chart)")
```
