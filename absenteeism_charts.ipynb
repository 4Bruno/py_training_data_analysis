```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import colormaps
import seaborn as sns
sns.set()

DIR_RESOURCES = 'data_analyst\Resources\Data'
DIR_EXPORTS = 'data_analyst\Exports'
dir_sources = os.path.join(os.getcwd(),DIR_RESOURCES)
dir_exports = os.path.join(os.getcwd(),DIR_EXPORTS)
abs_file = os.path.join(dir_sources,'Absenteeism-data.csv')
```

```python
def save_chart(name):
    if not '.png' in name:
        name += '.png'
    plt.savefig(os.path.join(dir_exports,name))
```

```python
df = pd.read_pickle(os.path.join(dir_exports,'absenteeism_cleaned.pkl'))
```

```python
# bar chart
%matplotlib inline
col_abs = df['Absenteeism Time in Hours']
col_edu = df['Education']
categories = ['High school','Bachelor Degree/Doctor']
plt.figure(figsize=(9,6))
plt.bar(x=df['Education'],height=col_abs,align='center',color="MidnightBlue")
plt.xticks([0,1],categories,rotation=0)
plt.title("Absenteeism based on education level", fontsize = 16, fontweight="bold")
save_chart('absenteeism_based_on_edu_level')
plt.show()
```
